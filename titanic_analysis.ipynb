## Tratamento de Duplicados

Linhas duplicadas podem distorcer a Análise Estatística e gerar interpretações incorretas, pois representam a repetição de um mesmo registro.

Por essa razão, precisamos identificá-las e eliminá-las.

```python
print(f"Linhas antes da remoção de duplicatas: {df_principais.shape[0]}")
df_principais_sem_duplicatas = df_principais.drop_duplicates().copy()
print(f"Linhas depois da remoção de duplicatas: {df_principais_sem_duplicatas.shape[0]}")
```

## Tratamento de Valores Nulos

Valores ausentes podem prejudicar a Análise, pois algumas técnicas estatísticas e visualizações dependem de informações completas.

Primeiro, identificamos estes casos e, após, decidimos qual a melhor estratégia para tratar esses valores.

```python
print('Valores nulos por coluna antes do tratamento(contagem):')
print(df_principais.isnull().sum())
print('Valores nulos por coluna antes do tratamento(porcentagem):')
print(df_principais.isnull().sum() / len(df_principais) * 100)
```

Algumas variáveis possuem valores ausentes, principalmente na coluna Age, que contém registros sem informação de idade. A coluna Embarked também apresenta uma pequena quantidade de valores nulos.

Para evitar a perda de muitos dados, os valores numéricos ausentes da coluna Age podem ser substituídos pela mediana, e os valores faltantes em Embarked podem ser preenchidos pela moda, por ser uma variável categórica.

```python
# Tratar a coluna 'Age' com a mediana
median_age = df_principais['Age'].median()
df_principais['Age'] = df_principais['Age'].fillna(median_age)
print("Valores nulos em 'Age' preenchidos com a mediana: {median_age}")

# Tratar a coluna 'Embarked' com a moda
mode_embarked = df_principais['Embarked'].mode()[0]
df_principais['Embarked'] = df_principais['Embarked'].fillna(mode_embarked)
print("Valores nulos em 'Embarked' preenchidos com a moda: {mode_embarked}")

# Verificar se tratamos todos os valores nulos
print("Valores nulos após o tratamento:")
print(df_principais.isnull().sum())
```

## Feature Engineering

Nesta etapa, criamos e transformamos variáveis a partir das informações do DataSet.

Deste modo, são geradas novas características que facilitam a Análise e mostram padrões relevantes relacionados à sobrevivência dos passageiros.

### Variável Tamanho da Família

Uma informação relevante é o tamanho da família do passageiro a bordo do Titanic. Essa informação pode ser obtida a partir das variáveis SibSp (número de irmãos/cônjuges) e Parch (número de pais/filhos).

A nova variável FamilySize será criada somando essas duas variáveis e adicionando o próprio passageiro.

```python
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
df[['SibSp','Parch','FamilySize']].head()
```

A nova variável FamilySize representa o número total de membros da família do passageiro presentes no navio. Com ela, podemos analisar se passageiros que viajavam sozinhos ou acompanhados tiveram as mesmas chances de sobrevivência.

### Variável Passageiro Sozinho

Com base no tamanho da família, identificamos se o passageiro estava viajando sozinho ou acompanhado. Para isso, criamos a variável IsAlone, que indicará se o passageiro estava sozinho no navio.

```python
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
df[['FamilySize','IsAlone']].head()
```

O valor 1 representa passageiros que estavam sozinhos e 0 indica que o passageiro estava com familiares.

Assim como a variável FamilySize, com a variável IsAlone podemos analisar se passageiros que viajavam sozinhos ou acompanhados tiveram as mesmas chances de sobrevivência.

### Grupos de Faixas Etárias


Para facilitar a análise da variável Age, as idades podem ser agrupadas em faixas etárias, permitindo analisar de forma mais clara como a idade pode ter influenciado as chances de sobrevivência.

## Sobrevivência por Gênero

Um dos fatores que podem ter influenciado a sobrevivência dos passageiros do Titanic é o gênero.

Há relatos históricos indicando que houve uma prioridade para mulheres e crianças no acesso aos botes salva-vidas.

Dessa forma, analisar a relação entre gênero e sobrevivência pode ajudar a identificar se esse fator teve impacto significativo nas chances de sobrevivência dos passageiros.

```python
# Contagem de sobreviventes por gênero
df.groupby('Sex')['Survived'].sum()

# Taxa de sobrevivência por gênero
df.groupby('Sex')['Survived'].mean()

# Visualização gráfica
sns.countplot(x='Sex', hue='Survived', data=df)

plt.title('Sobrevivência por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Número de Passageiros')
plt.legend(title='Sobreviveu', labels=['Não', 'Sim'])

plt.show()
```

Percebemos uma diferença significativa nas taxas de sobrevivência entre os gêneros, tendo as passageiras uma taxa de sobrevivência muito maior em comparação aos passageiros.

Esse padrão pode estar relacionado **às políticas adotadas durante o resgate**, que priorizavam mulheres e crianças no embarque nos botes salva-vidas.

Essa observação indica que o gênero foi um dos fatores mais relevantes para a sobrevivência no desastre do Titanic.

```python
# CÓDIGO ANTIGO -  COMPARAR - UTILIZAR GRAFICO BARPLOT??
#Sobrevivência por gênero

taxa_sobrevivencia_genero = df_principais.groupby('Sex')['Survived'].mean() * 100
print("Taxa de sobrevivência por gênero:")
display(taxa_sobrevivencia_genero)

# Gráfico de barras para sobrevivência por gênero
plt.figure(figsize=(6, 4))
sns.barplot(x=taxa_sobrevivencia_genero.index, y=taxa_sobrevivencia_genero.values, palette='viridis', hue=taxa_sobrevivencia_genero.index, legend=False)
plt.title('Taxa de Sobrevivência por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Taxa de Sobrevivência (%)')
plt.ylim(0, 100)
plt.show()
```

```python
## CÓDIGO ANTIGO - COMPARAR
#Idade vs Sobrevivência
#Calcular a média da idade dos passageiros que sobreviveram.

sobreviventes = df_principais[df_principais['Survived'] == 1]
media_idade_sobreviventes = sobreviventes['Age'].mean()

print(f"Média de idade dos sobreviventes: {media_idade_sobreviventes:.2f}")

#Sobrevivencia por Faixa Etária
#Criar coluna 'Faixa_Etaria' categorizando a coluna 'Age' em faixas etárias
#Calcular a taxa de sobrevivência para cada grupo, agrupando por 'Faixa_Etaria' e a média de 'Survived'

bins = [0, 12, 18, 40, 60, np.inf]
labels = ['0-12', '13-18', '19-40', '41-60', '60+']
df_principais['Faixa_Etaria'] = pd.cut(df_principais['Age'], bins=bins, labels=labels, right=False)

taxa_sobrevivencia_faixa_etaria = df_principais.groupby('Faixa_Etaria', observed=False)['Survived'].mean() * 100
print("Taxa de sobrevivência por faixa etária:")
display(taxa_sobrevivencia_faixa_etaria)

#Histograma de idade

plt.figure(figsize=(10, 6))
sns.histplot(df_principais['Age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Distribuição de Idade dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Contagem de Passageiros')
plt.show()

#Visualização da Sobrevivência por Faixa Etária

plt.figure(figsize=(10, 6))
sns.barplot(x=taxa_sobrevivencia_faixa_etaria.index, y=taxa_sobrevivencia_faixa_etaria.values, palette='viridis', hue=taxa_sobrevivencia_faixa_etaria.index, legend=False)
plt.title('Taxa de Sobrevivência por Faixa Etária')
plt.xlabel('Faixa Etária')
plt.ylabel('Taxa de Sobrevivência (%)')
plt.ylim(0, 100)
plt.show()
```

Essa combinação é particularmente interessante, pois permite observar se a diferença de sobrevivência entre homens e mulheres também varia de acordo com a classe do passageiro.

```python
sns.catplot(x='Pclass', hue='Survived', col='Sex', kind='count', data=df)

plt.show()
```

A análise mostra que mulheres apresentaram taxas de sobrevivência maiores em todas as classes sociais. No entanto, passageiros da primeira classe tiveram maiores chances de sobreviver em comparação com as classes mais baixas.

```python
# CÓDIGO ANTIGO - COMPARAR
#Mulheres da 3ª classe sobreviveram menos que da 1ª?
#Calcular Sobrevivência de Mulheres por Classe
#Filtrar o DataFrame para incluir apenas mulheres e, em seguida, calcular a taxa de sobrevivência média para cada Pclass (1ª, 2ª e 3ª classe) entre elas.
#First, I will filter the main DataFrame (df_principais) to select only female passengers. This filtered DataFrame will be stored in df_mulheres as requested, which is the initial step to analyze women's survival rates.
df_mulheres = df_principais[df_principais['Sex'] == 'female'].copy()
print("DataFrame com apenas passageiras do sexo feminino (df_mulheres) criado.")
# display(df_mulheres.head())

#Now that the DataFrame containing only female passengers (df_mulheres) is ready, I will calculate the survival rate for women across different social classes by grouping df_mulheres by 'Pclass' and computing the mean of the 'Survived' column, then convert it to percentage as requested.

taxa_sobrevivencia_mulheres_por_classe = df_mulheres.groupby('Pclass')['Survived'].mean() * 100
print("Taxa de sobrevivência de mulheres por classe social:")
display(taxa_sobrevivencia_mulheres_por_classe)

#To visualize the survival rate of women by social class, I will create a bar chart using seaborn.barplot with the calculated taxa_sobrevivencia_mulheres_por_classe. This will clearly show the survival differences across classes.

plt.figure(figsize=(8, 5))
sns.barplot(x=taxa_sobrevivencia_mulheres_por_classe.index, y=taxa_sobrevivencia_mulheres_por_classe.values, palette='pink', hue=taxa_sobrevivencia_mulheres_por_classe.index, legend=False)
plt.title('Taxa de Sobrevivência de Mulheres por Classe Social')
plt.xlabel('Classe do Passageiro (1=Primeira, 2=Segunda, 3=Terceira)')
plt.ylabel('Taxa de Sobrevivência (%)')
plt.ylim(0, 100)
plt.xticks(ticks=[0, 1, 2], labels=['1ª Classe', '2ª Classe', '3ª Classe'])
plt.show()
```

## Matriz de Correlação entre Variáveis Numéricas

Além das análises visuais, também é útil analisar a correlação entre variáveis numéricas. A matriz de correlação permite identificar relações entre diferentes variáveis e verificar se algumas delas apresentam associação com a variável de sobrevivência.

Além das análises gráficas realizadas anteriormente, também é útil examinar a correlação entre variáveis numéricas do dataset.

A matriz de correlação permite identificar relações entre diferentes variáveis, indicando se elas apresentam associação positiva, negativa ou pouca relação entre si.

Para facilitar a visualização dessas relações, será utilizado um heatmap, que representa graficamente os coeficientes de correlação entre as variáveis.

```python
# selecionar variáveis numéricas relevantes
colunas_correlacao = ['Age', 'Fare', 'SibSp', 'Parch', 'Survived']

# calcular matriz de correlação
matriz_correlacao = df[colunas_correlacao].corr()

# visualizar matriz
print("Matriz de Correlação:")
display(matriz_correlacao)

# heatmap
plt.figure(figsize=(8,6))
sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm', fmt=".2f")

plt.title('Matriz de Correlação entre Variáveis Numéricas')
plt.show()
```

A matriz de correlação permite observar o grau de relação entre as variáveis numéricas do dataset. Os valores variam entre -1 e 1, onde valores próximos de 1 indicam correlação positiva forte, valores próximos de -1 indicam correlação negativa forte e valores próximos de 0 indicam pouca ou nenhuma correlação.

Observa-se que algumas variáveis apresentam correlações moderadas, enquanto outras possuem pouca relação direta com a variável de sobrevivência. Isso sugere que a sobrevivência dos passageiros do Titanic provavelmente foi influenciada por uma combinação de fatores, como gênero, classe social e idade, em vez de uma única variável isolada.

```python

# CÓDIGO ANTIGO - COMPARAR

plt.figure(figsize=(8,6))

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')

plt.title('Matriz de Correlação das Variáveis Numéricas')

plt.show()

colunas_para_correlacao = ['Age', 'Fare', 'SibSp', 'Parch', 'Survived']
df_correlacao = df_principais[colunas_para_correlacao]
matriz_correlacao = df_correlacao.corr()

print("Matriz de Correlação:")
display(matriz_correlacao)

plt.figure(figsize=(8, 6))
sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Heatmap da Matriz de Correlação entre Variáveis Selecionadas')
plt.show()
```

A matriz de correlação mostra o grau de relação entre as variáveis numéricas do dataset. Observa-se que algumas variáveis apresentam correlações moderadas, enquanto outras possuem pouca relação direta entre si.

Embora nenhuma variável isolada apresente correlação extremamente alta com a sobrevivência, a análise conjunta de diferentes fatores — como classe social, gênero e idade — ajuda a explicar melhor as diferenças nas taxas de sobrevivência observadas.
